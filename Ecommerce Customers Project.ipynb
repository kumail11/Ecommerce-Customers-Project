```python
x = customer_data[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = customer_data['Yearly Amount Spent']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.2, random_state = 14)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
print(linear_model.coef_)
```

```python
y_pred = linear_model.predict(x_test)
```

```python
new_frame = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
new_frame
```

```python
fig = plt.figure(figsize = (12, 8))
new_frame.hist(bins = 50)
```

```python
fig = plt.figure(figsize = (12, 8))
plt.scatter(y_test, y_pred, color = 'red')
plt.show()
```

```python
fig = plt.figure(figsize = (12, 8))
sns.lmplot(x = 'Actual', y = 'Predicted', data = new_frame)
plt.show()
```

```python
print('MAE:', skm.mean_absolute_error(y_test, y_pred))
print('MSE:', skm.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(skm.mean_squared_error(y_test, y_pred)))
```

```python
fig = plt.figure(figsize = (12, 8))
sns.distplot((y_test-y_pred),bins=50)
```

```python
coeffecients = pd.DataFrame(linear_model.coef_, x.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
accuracy = linear_model.score(x_test, y_test)
```

```python
print('Accuracy of Model:', accuracy * 100, '%')
```
